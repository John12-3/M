```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.svm import SVC
```

```python
data = pd.read_csv("/content/pima-diabetes.csv")
data
```

```python
#  we need to have x and y
x = data.loc[:,data.columns != 'Outcome']
x
```

```python
y = data["Outcome"].values
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
# we need to do fit with linear first
from sklearn import svm
from sklearn.svm import SVC
linear = svm.SVC(kernel='linear')
linear.fit(x_train,y_train)
```

```python
# y_pred using the linear thing
y_pred = linear.predict(x_test)
```

```python

# check for the accuracy score , confusion matrix and classification report
print("Accuracy is ",accuracy_score(y_test,y_pred))
print("Confusion matrix is ",confusion_matrix(y_test,y_pred))
print("Classification report is ",classification_report(y_test,y_pred))
```

```python
# we need to do for other models as well - > sigmoid , poly and rbf
model1 = svm.SVC(kernel='sigmoid')
model2 = svm.SVC(kernel='poly')
model3 = svm.SVC(kernel='rbf')
```

```python
#  we need to fit it according to the model with the train datas
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)
```

```python
# we need to have the y_pred which will be different for each model
y_pred1 = model1.predict(x_test)
y_pred2 = model2.predict(x_test)
y_pred3 = model3.predict(x_test)
```

```python
print("Prediction by model1 : ",accuracy_score(y_test,y_pred1))
print("Prediction by model2 : ",accuracy_score(y_test,y_pred2))
print("Prediction by model3 : ",accuracy_score(y_test,y_pred3))
```
